```python
import pandas as pd
```

```python
df = pd.concat([
    pd.read_csv("../dataset/baseline/all-raw_scores.csv"),
    pd.read_csv("../dataset/baseline/manual-raw_scores.csv"),
    pd.read_csv("../dataset/baseline/pvalue-raw_scores.csv"),
    pd.read_csv("../dataset/baseline/r-raw_scores.csv"),
], ignore_index=True)

df.head()
```

```python
df_proj = df[df["name"] != "LinearRegression"]

df_p = df_proj.pivot_table(index=["run_name", "category", "name"], values=["fit_time", "features_count", "r2", "mae", "mape", "mse"], aggfunc="mean")
df_p.reset_index().to_csv("summary.csv", index=None)

df_p.sort_values(by="mse", ascending=True) \
    .style.background_gradient(cmap="Reds", subset=["fit_time", "mae", "mape", "mse"]) \
    .background_gradient(cmap="Blues", subset=["r2"])
```

```python
df_ratio = df_proj.pivot_table(index=["run_name", "category", "name"], values=["fit_time", "features_count", "r2", "mae", "mape", "mse"], aggfunc="mean") \
    .reset_index() \
    .sort_values(by="mse", ascending=True)


df_ratio = df_ratio[df_ratio["name"].isin(["RandomForestRegressor", "CatBoostRegressor"])]
df_ratio["features_mae_ratio"] = df_ratio["features_count"] / df_ratio["mae"]

df_ratio.sort_values(by="features_mae_ratio", ascending=False) \
    .style.background_gradient(cmap="Reds", subset=["fit_time", "mae", "mape", "mse"]) \
    .background_gradient(cmap="Blues", subset=["r2"]) \
    .background_gradient(cmap="Greens", subset=["features_mae_ratio"])
```
