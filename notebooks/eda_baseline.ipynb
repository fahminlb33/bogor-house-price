```python
import pandas as pd
import pyarrow.parquet
```

```python
df = pd.concat([
    pd.read_csv("../dataset/train_baseline/sel_all/raw_regression_scores.csv").assign(method="all", features_count=len(pyarrow.parquet.read_schema("../dataset/etl/L3.regression_train-sel_all.parquet", memory_map=True).names) - 2),
    pd.read_csv("../dataset/train_baseline/sel_manual/raw_regression_scores.csv").assign(method="manual", features_count=len(pyarrow.parquet.read_schema("../dataset/etl/L3.regression_train-sel_manual.parquet", memory_map=True).names) - 2),
    pd.read_csv("../dataset/train_baseline/sel_r/raw_regression_scores.csv").assign(method="r", features_count=len(pyarrow.parquet.read_schema("../dataset/etl/L3.regression_train-sel_r.parquet", memory_map=True).names) - 2),
    pd.read_csv("../dataset/train_baseline/sel_pval/raw_regression_scores.csv").assign(method="pval", features_count=len(pyarrow.parquet.read_schema("../dataset/etl/L3.regression_train-sel_pval.parquet", memory_map=True).names) - 2),
], ignore_index=True)

df.head()
```

```python
rdf = df.pivot_table(index=["method", "category", "name"], values=["fit_time", "features_count", "r2", "mse", "mae", "mape"], aggfunc="mean")
rdf.sort_values(by="mse", ascending=True).style.background_gradient(cmap="Reds", subset=["fit_time", "mae", "mape", "mse"]).background_gradient(cmap="Blues", subset=["r2"])
```
